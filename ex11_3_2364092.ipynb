```python
class Player:
    def __init__(self, number, name, profile):
        self.number = number
        self.name = name
        self.profile = profile

def senkeilist(filename):
    players = []
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            data = line.split()
            player = Player(int(data[0]), data[1], ' '.join(data[2:]))
            players.insert(0, player)
    return players

def outputlist(players):
    for player in players:
        print(f"{player.number:2d} {player.name} {player.profile}")

def delete_player(players, number):
    for i in range(len(players)):
        if players[i].number == number:
            print("---削除しました---")
            del players[i]
            return
    print("---該当者はいません---")

def main():
    players = senkeilist("Zac_Japan.txt")
    outputlist(players)

    while True:
        number = int(input("削除する登録番号は？ "))
        if number == 0:
            print("プログラムを終了します")
            break

        delete_player(players, number)
        outputlist(players)

main()
```
